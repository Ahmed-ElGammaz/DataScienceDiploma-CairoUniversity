```python
import numpy as np
```

```python
A = np.array([[2,-3],[4,4]]) # matrix A(2,2)
A
```

```python
A_inverse = np.linalg.inv(A)   # inverse matrix A
A_inverse
```

```python
B = np.array([[3,1],[5,2]])   # matrix B(2,2)
B
```

```python
B_inverse = np.linalg.inv(B)   # inverse matrix B
B_inverse
```
